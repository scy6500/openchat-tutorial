# API 통신에 의한 openchat 사용

```python
# API 통신을 위해 requests 모듈 import
import requests
```

```python
# 대화를 종료하려면 .exit를 입력하세요.
text = input("user :   ")
while text != ".exit":
  response = requests.post("https://main-openchat-fpem123.endpoint.ainize.ai/send/API_user", data={'text': text})
  status_code = response.status_code
  if status_code == 200:
    print("bot : ", response.json()["output"])
    text = input("user :   ")
  elif status_code == 400:
    print("Bad Request Error")
    break
  elif status_code == 429:
    print("Too many requests")
    break
  elif status_code == 500:
    print("Server-side Error")
    break
```
