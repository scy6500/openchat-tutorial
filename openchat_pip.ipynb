## Run openchat

### Supported Models
**GPT-Neo**
* gptneo.small
* gptneo.medium
* gptneo.large
* gptneo.xlarge

**Blender**
* blender.small
* blender.medium
* blender.large
* blender.xlarge
* blender.xxlarge

**DialoGPT**
* dialogpt.small
* dialogpt.medium
* dialogpt.large

Use these names as parameter model='name' when you create OpenChat.

For example, if you want to use blender large -> OpenChat(model="blender.large")
