# pip install을 통한 openchat 사용

### openchat 실행(1.0 version 기준입니다.)

model 종류 : dialogpt, blenderbot

dialogpt size 종류 : small, medium, large

blenderbot size 종류 : small, medium, large, xlarge

모델의 종류나 사이즈 변경할 시 파라미터를 원하는 것으로 변경하면 됨

ex) blenderbot large를 사용하고 싶다 -> OpenChat(model="blenderbot", size="large", device="cuda")

```python
from openchat import OpenChat
```

```python
# 상단에 런타임 -> 런타임 유형 변경 -> 하드웨어 가속기 : gpu 설정 후 openchat 실행
# 대화 종료를 원하면 .exit 입력
OpenChat(model="blenderbot", size="large", device="cuda")
```

```python
# gpu 설정이 불가하면 해당 블록으로 openchat 실행
# 대화 종료를 원하면 .exit 입력
OpenChat(model="blenderbot", size="large")
```
